```python
# Import library yang akan digunakan
# pandas digunakan untuk membaca dataset
# pathlib digunakan untuk mengelola path file

import pandas as pd
from pathlib import Path
```

```python
# Menentukan lokasi project
# BASE_DIR = folder utama project
# RAW_DIR = folder tempat data mentah disimpan

BASE_DIR = Path.cwd().parent
RAW_DIR = BASE_DIR / "data" / "raw"

print(BASE_DIR)
print(RAW_DIR)
```

```python
# Melihat folder dataset yang tersedia

for folder in RAW_DIR.iterdir():
    print(folder.name)
```

```python
# Melihat file yang tersedia pada setiap folder

for folder in RAW_DIR.iterdir():

    print(f"\n=== {folder.name} ===")

    for file in folder.glob("*"):
        print(file.name)
```
