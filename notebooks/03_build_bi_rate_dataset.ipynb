```python
# Import library yang dibutuhkan
# pandas digunakan untuk membaca dan mengolah data tabel
# pathlib digunakan untuk mengatur path folder/file

import pandas as pd
from pathlib import Path
```

```python
# Menentukan lokasi folder project
# BASE_DIR = folder utama project
# RAW_DIR = folder tempat data mentah disimpan
# PROCESSED_DIR = folder tempat data hasil preprocessing disimpan
# BI_RATE_DIR = folder khusus data BI Rate

BASE_DIR = Path.cwd().parent

RAW_DIR = BASE_DIR / "data" / "raw"
PROCESSED_DIR = BASE_DIR / "data" / "processed"
BI_RATE_DIR = RAW_DIR / "BI-Rate"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

print("Base directory:", BASE_DIR)
print("BI Rate directory:", BI_RATE_DIR)
```

```python
# Mengambil nama bulan dari baris ke-2
# Kolom 1 sampai 12 berisi Januari sampai Desember
# Kolom 13 adalah Tahunan, jadi tidak digunakan

month_names = bi_df.iloc[2, 1:13].tolist()

# Mengambil nilai BI Rate dari baris ke-3
# Kolom 1 sampai 12 berisi nilai BI Rate bulanan

values = bi_df.iloc[3, 1:13].tolist()

print(month_names)
print(values)
```

```python
# Mapping nama bulan Indonesia ke angka bulan
# Ini digunakan untuk membuat format tanggal standar

months = {
    "Januari": "01",
    "Februari": "02",
    "Maret": "03",
    "April": "04",
    "Mei": "05",
    "Juni": "06",
    "Juli": "07",
    "Agustus": "08",
    "September": "09",
    "Oktober": "10",
    "November": "11",
    "Desember": "12",
}
```

```python
# Mengubah data BI Rate tahun 2010 menjadi format time series bulanan

year = 2010

bi_rate_2010 = pd.DataFrame({
    "date": [f"{year}-{months[month]}-01" for month in month_names],
    "bi_rate": values
})

# Mengubah tipe data date menjadi datetime
bi_rate_2010["date"] = pd.to_datetime(bi_rate_2010["date"])

# Mengubah BI Rate menjadi numeric
bi_rate_2010["bi_rate"] = pd.to_numeric(
    bi_rate_2010["bi_rate"],
    errors="coerce"
)

bi_rate_2010
```

```python
# Function untuk membangun dataset BI Rate per tahun
# Input: tahun
# Output: dataframe date dan bi_rate untuk 12 bulan pada tahun tersebut

def build_bi_rate_year(year):
    # Membaca file BI Rate berdasarkan tahun
    file_path = BI_RATE_DIR / f"BI Rate, {year}.csv"
    df = pd.read_csv(file_path)

    # Mengambil nama bulan dari baris ke-2
    month_names = df.iloc[2, 1:13].tolist()

    # Mengambil nilai BI Rate dari baris ke-3
    values = df.iloc[3, 1:13].tolist()

    # Mengubah format melebar menjadi format time series bulanan
    result = pd.DataFrame({
        "date": [f"{year}-{months[month]}-01" for month in month_names],
        "bi_rate": values
    })

    # Mengubah date menjadi datetime
    result["date"] = pd.to_datetime(result["date"])

    # Mengubah bi_rate menjadi angka
    result["bi_rate"] = pd.to_numeric(
        result["bi_rate"],
        errors="coerce"
    )

    return result
```

```python
# Menguji function pada tahun 2015

build_bi_rate_year(2015)
```

```python
# Menggabungkan seluruh data BI Rate dari tahun 2010 sampai 2025
# Setiap file tahunan akan diproses dengan function build_bi_rate_year()
# Hasil akhirnya adalah satu dataset BI Rate bulanan

bi_rate_all = pd.concat(
    [
        build_bi_rate_year(year)
        for year in range(2010, 2026)
    ],
    ignore_index=True
)

bi_rate_all
```

```python
# Menyimpan dataset BI Rate yang sudah rapi ke folder processed

output_path = PROCESSED_DIR / "bi_rate_monthly.csv"

bi_rate_all.to_csv(
    output_path,
    index=False
)

print(f"Dataset saved: {output_path}")
```
