```python
# Import library yang dibutuhkan
# pandas digunakan untuk membaca dan mengolah data tabel
# pathlib digunakan untuk mengatur path folder/file

import pandas as pd
from pathlib import Path
```

```python
# Menentukan lokasi folder project
# BASE_DIR = folder utama project
# RAW_DIR = folder tempat data mentah disimpan
# PROCESSED_DIR = folder tempat data hasil preprocessing disimpan

BASE_DIR = Path.cwd().parent

RAW_DIR = BASE_DIR / "data" / "raw"
PROCESSED_DIR = BASE_DIR / "data" / "processed"

EXCHANGE_RATE_DIR = RAW_DIR / "Exchange-Rate"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

print("Base directory:", BASE_DIR)
print("Exchange Rate directory:", EXCHANGE_RATE_DIR)
```

```python
# Membaca dataset exchange rate dari FRED
# Dataset ini sudah berbentuk time series bulanan

file_path = EXCHANGE_RATE_DIR / "CCUSMA02IDM618N.csv"

exchange_df = pd.read_csv(file_path)

exchange_df.head()
```

```python
# Mengecek ukuran dataset
# shape = (jumlah baris, jumlah kolom)

exchange_df.shape
```

```python
# Mengecek nama kolom dataset

exchange_df.columns.tolist()
```

```python
# Mengecek informasi tipe data awal

exchange_df.info()
```

```python
# Membuat salinan dataset agar data asli tetap aman

exchange_rate = exchange_df.copy()

exchange_rate.head()
```

```python
# Mengubah nama kolom agar lebih mudah dipahami
# observation_date -> date
# CCUSMA02IDM618N -> exchange_rate

exchange_rate = exchange_rate.rename(
    columns={
        "observation_date": "date",
        "CCUSMA02IDM618N": "exchange_rate"
    }
)

exchange_rate.head()
```

```python
# Mengubah kolom date menjadi datetime
# Mengubah exchange_rate menjadi numeric untuk memastikan nilainya angka

exchange_rate["date"] = pd.to_datetime(
    exchange_rate["date"],
    errors="coerce"
)

exchange_rate["exchange_rate"] = pd.to_numeric(
    exchange_rate["exchange_rate"],
    errors="coerce"
)

exchange_rate.info()
```

```python
# Mengecek ukuran dataset
# Harusnya 192 baris dan 2 kolom

exchange_rate.shape
```

```python
# Mengecek periode awal dan akhir data

print(exchange_rate["date"].min())
print(exchange_rate["date"].max())
```

```python
# Mengecek missing value

exchange_rate.isnull().sum()
```

```python
# Mengecek apakah ada tanggal duplikat

exchange_rate["date"].duplicated().sum()
```

```python
# Menyimpan dataset exchange rate yang sudah rapi ke folder processed

output_path = PROCESSED_DIR / "exchange_rate_monthly.csv"

exchange_rate.to_csv(
    output_path,
    index=False
)

print(f"Dataset saved: {output_path}")
```
